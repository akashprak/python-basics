#### 1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?
**Ans**: Probability distribution describes the probabilities of occurrence of different possible outcomes for an experiment.  
Based on the likelihood of the previous data, it can predict the probability of occurence of a value.


#### 2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?
**Ans**: Pseudo-random numbers are numbers which are generated by some deterministic process but they qualify the statistical test for randomness. They may be generated by using a truly random number.  
The latter is considered good enough because for most applications it is statistically significant.

#### 3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

#### 5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?
**Ans**: The actual probability gets closer to the exact probability as the number of trial grows, this is known as law of large numbers.
